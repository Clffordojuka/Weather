```python
#Weather was exactly clear
clear_weather = Data[Data['Weather'] == 'Clear']
clear_weather_records = clear_weather.shape[0]
clear_weather.head()
```

```python
#The number of times the windspeed was exactly 4km/h
wind_speed = Data[Data['Wind Speed_km/h'] == 4]
wind_speed_count = wind_speed.shape[0]
wind_speed_count
```

```python
#Checking for presence of null values
null_values = Data.isnull().sum()
null_values
```

```python
#Renaming column "Weather" to "Weather-Condition"
Data.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
Data.head()
```

```python
#Finding mean visibility of datasets
mean_visibility = Data['Visibility_km'].mean()
mean_visibility
```

```python
#Wind speed greater than 24km/h and visibility equal to 25km
wind_speed_visibility = Data[(Data['Wind Speed_km/h'] > 24) & (Data['Visibility_km'] == 25)]
wind_speed_visibility_count = wind_speed_visibility.shape[0]
wind_speed_visibility_count
```

```python
#Mean value of each column of weather condition
numeric_columns = Data.select_dtypes(include=['number']).columns
mean_values_per_weather = Data.groupby('Weather_Condition')[numeric_columns].mean()
mean_values_per_weather
```

```python
#All instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
specific_conditions = Data[(Data['Weather_Condition'] == 'Clear') & (Data['Rel Hum_%'] > 50) | (Data['Visibility_km'] > 40)]
specific_conditions.head()
```

```python
#The number of weather conditions that include snow.
snow_conditions = Data[Data['Weather_Condition'].str.contains('Snow', case=False)]
snow_conditions_count = snow_conditions.shape[0]
snow_conditions_count
```
